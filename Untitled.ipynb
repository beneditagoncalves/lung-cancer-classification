```python
# Plot the distribution of malignancy
plt.figure(figsize=(8, 6))
sns.histplot(df['malignancy'], bins=10, kde=True)
plt.title("Distribution of Malignancy")
plt.show()
```

The distribution of malignancy is not uniform, with many nodules having a malignancy rating of 3. This suggests some class imbalance.
We need to address this imbalance with techniques like oversampling, undersampling, or using class weights in your models to prevent bias toward more common classes.

```python
# Drop rows where malignancy is 3
df = df[df['malignancy'] != 3]

# Create a new binary 'is_malignant' column (1 for malignant, 0 for benign)
df['is_malignant'] = df['malignancy'].apply(lambda x: 1 if x >= 4 else 0)
```

```python
# Check the distribution of the new 'is_malignant' column
df['is_malignant'].value_counts()
```

This distribution indicates that the dataset is imbalanced, with significantly more benign cases than malignant cases. This imbalance could affect the performance of machine learning models, as they may become biased toward predicting the majority class (benign) more frequently.

```python
# Split into input features (X) and target (y)
X = df.drop(columns=['malignancy', 'is_malignant'])  # Drop the original malignancy column and target
y = df['is_malignant']  # Binary target (0 for benign, 1 for malignant)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Apply SMOTE to oversample the minority class in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Train the model on the SMOTE-resampled dataset
model_smote = RandomForestClassifier(random_state=42)
model_smote.fit(X_train_smote, y_train_smote)

# Predict on the test set
y_pred_smote = model_smote.predict(X_test)

# Evaluate the performance
print("SMOTE - Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred_smote))
print("SMOTE - F1 Score:", f1_score(y_test, y_pred_smote))
print("SMOTE - ROC-AUC:", roc_auc_score(y_test, model_smote.predict_proba(X_test)[:, 1]))
```

SMOTE got an higher balanced accuracy and a slightly better F1 score, which suggests that it is handling the minority class better than Class Weights.
Class Weights has a slightly better ROC-AUC, but the difference is small and given the importance of the balanced accuracy and F1 score in imbalanced datasets, we decided to go with SMOTE.
